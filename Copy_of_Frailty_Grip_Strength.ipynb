```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 2: Upload the dataset
uploaded = files.upload()

# Step 3: Read the uploaded file
frailty_data = pd.read_csv('frailty_data.csv')

# Step 4: Display first few rows
print("First 5 rows of Frailty Dataset:")
print(frailty_data.head())

# Step 5: Check for missing values
print("\nMissing Values in Dataset:")
print(frailty_data.isnull().sum())

# Step 6 (Option 1): Use Correct Column Name
# If you don't want to modify the column names, just use "Grip Strength"
correlation = frailty_data['Grip Strength'].corr(frailty_data['Frailty'].apply(lambda x: 1 if x == 'Y' else 0))
print(f"\nCorrelation between Grip Strength and Frailty: {correlation:.4f}")

# Step 6 (Option 2): Standardize Column Names to Lowercase
frailty_data.columns = frailty_data.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 7: Convert 'frailty' to binary (1 for 'Y', 0 for 'N')
frailty_data['frailty_binary'] = frailty_data['frailty'].apply(lambda x: 1 if x == 'Y' else 0)

# Step 8: Compute Correlation Using Standardized Column Names
correlation = frailty_data['grip_strength'].corr(frailty_data['frailty_binary'])
print(f"\nCorrelation between Grip Strength and Frailty: {correlation:.4f}")

# Step 9: Plot Grip Strength vs. Frailty
plt.figure(figsize=(8, 6))
sns.boxplot(x='frailty', y='grip_strength', data=frailty_data)
plt.title("Grip Strength vs. Frailty")
plt.xlabel("Frailty (No = 0, Yes = 1)")
plt.ylabel("Grip Strength (kg)")
plt.show()

# Step 10: Save the plot (optional)
plt.savefig('grip_strength_vs_frailty.png')
```
