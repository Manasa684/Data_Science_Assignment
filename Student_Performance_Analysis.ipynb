```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 2: Upload the dataset
uploaded = files.upload()

# Step 3: Read the uploaded file
student_data = pd.read_csv('StudentsPerformance.csv')

# Step 4: Display first few rows
print("First 5 rows of Student Performance Dataset:")
print(student_data.head())

# Step 5: Create Data Visualizations

# Visualization 1: Math Scores Distribution
plt.figure(figsize=(8, 6))
sns.histplot(student_data['math score'], bins=20, kde=True, color='blue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
plt.savefig('math_scores_distribution.png')

# Visualization 2: Reading Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=student_data, palette='Set2')
plt.title("Reading Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Reading Score")
plt.show()
plt.savefig('reading_scores_by_gender.png')

# Visualization 3: Writing Scores by Lunch Type
plt.figure(figsize=(8, 6))
sns.boxplot(x='lunch', y='writing score', data=student_data, palette='Set3')
plt.title("Writing Scores by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Writing Score")
plt.show()
plt.savefig('writing_scores_by_lunch.png')

# Visualization 4: Math Scores by Parental Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=student_data, palette='Set1')
plt.title("Math Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Math Score")
plt.xticks(rotation=45)
plt.show()
plt.savefig('math_scores_by_parental_education.png')

# Visualization 5: Test Preparation Course Impact on Scores
plt.figure(figsize=(12, 6))
sns.boxplot(x='test preparation course', y='math score', data=student_data, palette='Set2')
plt.title("Test Preparation Course Impact on Math Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()
plt.savefig('test_preparation_impact_math.png')

plt.figure(figsize=(12, 6))
sns.boxplot(x='test preparation course', y='reading score', data=student_data, palette='Set2')
plt.title("Test Preparation Course Impact on Reading Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Reading Score")
plt.show()
plt.savefig('test_preparation_impact_reading.png')

plt.figure(figsize=(12, 6))
sns.boxplot(x='test preparation course', y='writing score', data=student_data, palette='Set2')
plt.title("Test Preparation Course Impact on Writing Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Writing Score")
plt.show()
plt.savefig('test_preparation_impact_writing.png')
```
